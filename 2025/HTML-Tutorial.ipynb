After completing this tutorial, you should be capable to:

-create a dropdown menu with various options,

We beginn each HTML document by declaring the document type. Furthermore HTML documents use opening- and closing-tags to define the boundaries of differnt element. e.g. < html> is the opening- and </ html> the corresponding closing-tag. We usally want to define our elements inside the html element. Elements declared inside another element are called child-element. The following example shows a head elemten with a title element inside. The title will later be shown at the top of the browser. (double click next line to see code)

<!DOCTYPE html>
<html>          
    <head>
        <title>This is a Title</title>
    </head>
</html> 

We use the body element to generate everything that is supposed to be shown on the website. Inside these bodies we define the elements we want to use like text, tables, dropdown menus, images, hyperlinks and so on. The following example show the insertion of an image. Notice that the img tag does not use a closing tag. Some elements cannot have a child-element ans thus require no closing-tag. (double click next line to see code)

<!DOCTYPE html>
<html>          
    <head>
        <title>This is a Title</title>
    </head>
    <body>
        <h1>This is Header</h1>
        <p>This is a paragraph.</p>
        <img src= "./images/LogoAI.png">
    </body>
</html> 

Another important tag is the div-tag. We use divs to sepperate different parts on our website. This way we can seperate elements and later on modify for exapmle the position, color or appaerance of single elements. The following exaple shows a dropdown menu.

Notice that the dropdown menu is created by using a select-tag. We give this tag an "id" attribute which will allow us to reference this specific tag later. Each option-tag represents an option in our dropdown menu. (double click next line to see code)

<!DOCTYPE html>
<html>          
    <head>
        <title>This is a Title</title>
    </head>
    <body>
        <div>
           <select name="Dropdown Menu" id="menu">
              <option value="option 1">Option 1</option>
              <option value="option 2">Option 2</option>
              <option value="option 3">Option 3</option>
           </select>
        </div> 
    </body>
</html>
